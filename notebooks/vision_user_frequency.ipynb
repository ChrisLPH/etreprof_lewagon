# Frequency of user usage

## Imports

```python
import pandas as pd
import numpy as np
```

## Create dataframe

Generate 1000 random user ids for demo

```python
random_ids = np.random.choice(range(1, 100001), size=1000, replace=False)
```

Create Dataframe

```python
df_users_frequency = pd.DataFrame({
    'id': random_ids
})
```

Create columns

```python
# Add subscribe date
df_users_frequency['join_date'] = 0

# Total number of interactions
df_users_frequency['total_interactions'] = 0

# Add 12 columns for the last 12 weeks
for i in range(12):
    df_users_frequency[f'week_minus_{i}'] = 0

# Add 12 columns for the last 12 months
for i in range(12):
    df_users_frequency[f'month_minus_{i}'] = 0

# Add 3 columns for the last 3 years
for i in range(3):
    df_users_frequency[f'year_minus_{i}'] = 0
```
